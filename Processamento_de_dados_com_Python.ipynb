**Processamento de dados**

---



Temos um arquivo de pacientes que desenvolveram diabetes ou não. Foi solicitado gerar uma amostra de dados com os pacientes com mais de 50 anos, e para cada um deles, indicar em uma nova coluna se o paciente está normal (BMI menor que 30), ou obeso (BMI maior ou igual que 30). Gerar um novo arquivo CSV e encaminhar ao solicitante.

```python
import pandas as pd
import numpy as np
```

```python
#Leitura de arquivo CSV e criação de DataFrame.
df = pd.read_csv('diabetes.csv')
```

```python
#Visualização dos dados.
df.head()
```

```python
#Descripção dos dados.
df.info()
```

```python
df.describe()
```

```python
print('Dimensões do DataFrame: {}'.format(df.shape))
```

```python
#Verificando se existem valores nulos na df.
df.isnull().sum()
```

Dado que no existem valores nulos e não precisamos transformações de tipos nos dados, o proximo passo será criar a nova DataFrame.

```python
#DataFrame com pacientes com mais de 50 anos de idade.
novo_df = df[df['Age']>50]
```

```python
#Visualização dos dados.
novo_df.head()
```

```python
print('dimensões do novo DataFrame: {}' .format(novo_df.shape))
```


De um total de 786 registros, obtivemos uma amostra de 81 pacientes que atendem à condição de terem idade superior a 50 anos.

```python
#Adicionando a nova coluna para indocar se o paciente está normal ou obeso com base no BMI.
novo_df['Perfil']= np.where(novo_df['BMI']>=30, 'Obeso', 'Normal')
```

```python
#Visualização dos dados.
novo_df.head()
```

```python
#Criando o novo arquivo CSV com a informação solicitada.
novo_df.to_csv('novo_df.csv', index=False)
```
